```
file_path = '/content/drive/MyDrive/CSSI Notes /Hotelsss.csv'
hotels = pd.read_csv(file_path)
hotels.head()
```

```
hotels.count
```

```
missing_hotels = hotels['price'].isna()
missing_hotels.head()
```

```
cleaned_hotels = hotels.loc[(~missing_hotels).values, :]
```

```
cleaned_hotels['price'] = cleaned_hotels['price'].round(2)
```

```
cleaned_hotels.head()
```

```
cleaned_hotels.count
```

```
country_continent = cleaned_hotels[['continent', 'country']]
country_continent = country_continent.drop_duplicates()
country_continent
```

```
country_continent.to_csv("country_continent_data.csv", index=True)
cleaned_hotels.to_csv("cleaned_hotels_data.csv", index=True)
```
